```python
import astropy
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from astropy.io import ascii
from matplotlib.ticker import ScalarFormatter
import matplotlib.ticker as ticker

%matplotlib inline
data13mmsub = ascii.read("Per33.1.3mm.uv.radial_averaged.proj.30klam.txt",names=['uvd','amp','e_amp','real','e_real','imag','e_imag','number'] )
data13mm = ascii.read("../Per33/Per33.1.3mm.uv.radial_averaged.proj.30klam.txt",names=['uvd','amp','e_amp','real','e_real','imag','e_imag','number'] )

print data13mm
```

```python
fig=plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
ax.set_xlim(0,1500.0)
#ax.set_ylim(0.01,1000.0)
lin1=ax.plot(data13mm['uvd'],data13mm['real'],marker='o',color='black',linestyle='none')
lin1err=ax.errorbar(data13mm['uvd'],data13mm['real'], yerr=data13mm['e_real'], fmt='o',color='black')
lin2=ax.plot(data13mmsub['uvd'],data13mmsub['real'],marker='o',color='gray',linestyle='none')
lin2err=ax.errorbar(data13mmsub['uvd'],data13mmsub['real'], yerr=data13mmsub['e_real'], fmt='o',color='gray')

ax.set_title('1.3mm Real Visibility Amplitudes', fontsize=18)
ax.set_ylabel('Real Visibility Amplitude (Jy)', fontsize=18)
ax.set_xlabel('Projected uv-distance (Klambda)', fontsize=18)
```

```python
fig=plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
ax.set_xlim(0,1500.0)
ax.set_ylim(-0.10,0.15)
lin1=ax.plot(data13mm['uvd'],data13mm['real'],marker='o',color='black',linestyle='none')
lin1err=ax.errorbar(data13mm['uvd'],data13mm['real'], yerr=data13mm['e_real'], fmt='o',color='black')
lin2=ax.plot(data13mmsub['uvd'],data13mmsub['real'],marker='o',color='gray',linestyle='none')
lin2err=ax.errorbar(data13mmsub['uvd'],data13mmsub['real'], yerr=data13mmsub['e_real'], fmt='o',color='gray')

ax.set_title('1.3mm Real Visibility Amplitudes', fontsize=18)
ax.set_ylabel('Real Visibility Amplitude (Jy)', fontsize=18)
ax.set_xlabel('Projected uv-distance (Klambda)', fontsize=18)
```

```python
fig=plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
ax.set_xlim(0,1500.0)

lin1=ax.plot(data13mm['uvd'],data13mm['imag'],marker='o',color='black',linestyle='none')
lin1err=ax.errorbar(data13mm['uvd'],data13mm['imag'], yerr=data13mm['e_imag'], fmt='o',color='black')
lin2=ax.plot(data13mmsub['uvd'],data13mmsub['imag'],marker='o',color='gray',linestyle='none')
lin2err=ax.errorbar(data13mmsub['uvd'],data13mmsub['imag'], yerr=data13mmsub['e_imag'], fmt='o',color='gray')

ax.set_title('1.3mm Imaginary Visibility Amplitudes', fontsize=18)
ax.set_ylabel('Imaginary Visibility Amplitude (Jy)', fontsize=18)
ax.set_xlabel('Projected uv-distance (Klambda)', fontsize=18)
```

```python
fig=plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
ax.set_xlim(10,1500.0)
#ax.set_ylim(0.01,1000.0)
lin1=ax.loglog(data13mm['uvd'],data13mm['real'],marker='o',color='black',linestyle='none')
lin1err=ax.errorbar(data13mm['uvd'],data13mm['real'], yerr=data13mm['e_real'], fmt='o',color='black')
lin2=ax.loglog(data13mmsub['uvd'],data13mmsub['real'],marker='o',color='gray',linestyle='none')
lin2err=ax.errorbar(data13mmsub['uvd'],data13mmsub['real'], yerr=data13mmsub['e_real'], fmt='o',color='gray')

ax.set_title('1.3mm Real Visibility Amplitudes', fontsize=18)
ax.set_ylabel('Real Visibility Amplitude (Jy)', fontsize=18)
ax.set_xlabel('Projected uv-distance (Klambda)', fontsize=18)
for axis in [ax.xaxis, ax.yaxis]:
    axis.set_major_formatter(ScalarFormatter())
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y,pos: ('{{:.{:1d}f}}'.format(int(np.maximum(-np.log10(y),0)))).format(y)))
```
