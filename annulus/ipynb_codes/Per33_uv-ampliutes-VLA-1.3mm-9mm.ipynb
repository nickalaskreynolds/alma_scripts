```python
import astropy
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from astropy.io import ascii
from matplotlib.ticker import ScalarFormatter
import matplotlib.ticker as ticker

%matplotlib inline
data13mmsub = ascii.read("Per33.1.3mm.uv.radial_averaged.proj.30klam.txt",names=['uvd','amp','e_amp','real','e_real','imag','e_imag','number'] )
data9mmsub = ascii.read("Per33.9mm.uv.radial_averaged.proj.30klam.txt",names=['uvd','amp','e_amp','real','e_real','imag','e_imag','number'] )

print data13mmsub
print data9mmsub
```

```python
fig=plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
ax.set_xlim(0,1000.0)
#ax.set_ylim(-0.001,0.01)

scalefactor=data13mmsub['real'][5]/data9mmsub['real'][5]

print scalefactor
print data13mmsub['real'][3], data9mmsub['real'][3]
print data13mmsub['real'][4], data9mmsub['real'][4]

data9mmsub_scale=data9mmsub['real'].copy()*scalefactor
data9mmsub_scale_error=data9mmsub['e_real'].copy()*scalefactor
imag9mmsub_scale=data9mmsub['imag'].copy()*scalefactor
imag9mmsub_scale_error=data9mmsub['e_imag'].copy()*scalefactor


lin1=ax.plot(data9mmsub['uvd'],data9mmsub_scale,marker='o',color='black',linestyle='none')
lin1err=ax.errorbar(data9mmsub['uvd'],data9mmsub_scale, yerr=data9mmsub_scale_error, fmt='o',color='black')

lin2=ax.plot(data13mmsub['uvd'],data13mmsub['real'],marker='o',color='gray',linestyle='none')
lin2err=ax.errorbar(data13mmsub['uvd'],data13mmsub['real'], yerr=data13mmsub['e_real'], fmt='o',color='gray')






ax.set_title('8mm Real Visibility Amplitudes', fontsize=18)
ax.set_ylabel('Real Visibility Amplitude (Jy)', fontsize=18)
ax.set_xlabel('Projected uv-distance (Klambda)', fontsize=18)
```
